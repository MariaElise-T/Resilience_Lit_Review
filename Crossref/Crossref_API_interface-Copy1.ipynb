```python
import urllib.request, json 
```

Set up the query language and cursor

```python
base_call = 'https://api.crossref.org/works?query=resilient+resilience+resiliency&filter=has-abstract:true,type:journal-article&cursor='
json_arr = []
```

Pull data from crossref into data

```python
with urllib.request.urlopen(base_call+'*') as url:
    data = json.load(url)
cursor = data['message']['next-cursor']
json_arr.append(data)
results=1
i=0
while(results>0):
    with urllib.request.urlopen(base_call+cursor) as url:
        if(i % 100)==0:
            print("Run ", i, " completed.")
        data = json.load(url)
        cursor = data['message']['next-cursor']
        json_arr.append(data)
        results = len(data['message']['items'])
        i=i+1
```

```python
len(json_arr)
```

```python
crossref_list = list()
for page in json_arr: # cycle through pages of results
    page_clean = page['message']['items'] # eliminate page headers
    for record in page_clean: # add individual records
        crossref_list.append(record)
```

```python
len(crossref_list)
```

```python
crossref_list[0]
```

```python
include_list = list()
for item in crossref_list:
    if((('resilience' in item['title'][0].lower()) or ('resiliency' in item['title'][0].lower()) or ('resilient' in item['title'][0].lower())) and
    ((('resilience' in item['abstract'].lower()) or ('resiliency' in item['abstract'].lower()) or ('resilient' in item['abstract'].lower())))):
        include_list.append(item)
```

```python
len(include_list)
```

```python
final_json_crossref = json.dumps(include_list)
```

```python
with open("crossref_has_abstract.json", "w") as outfile:
    outfile.write(final_json_crossref)
```
