```python
import json
```

```python
f = open('/Users/maria/Desktop/Resilience_Lit_Review/crossref_has_abstract_cleaned.json')
data = json.load(f)
```

```python
data[0]
```

```python
top_ten = sorted(data, key=lambda x: x['is-referenced-by-count'])[-10:]
```

```python
top_ten_neat = list()
for item in top_ten:
    top_ten_small = {key: item[key] for key in ['DOI', 'abstract', 'is-referenced-by-count', 'title'] }
    top_ten_neat.append(top_ten_small)
```
