```python
import requests
import json
import pandas as pd
import numpy 
import csv

MY_API_KEY = '346e2d65ab4c7ef5217ca1de44cd06b3'
```

```python
scopus_df = pd.read_csv('/Users/maria/Desktop/scopus/scopus.csv')
```

```python
scopus_ids_np = scopus_df['scopus_id'].to_numpy()
```

```python
scopus_ids_np_clean = {x.replace('SCOPUS_ID:', '') for x in scopus_ids_np}
```

```python
#https://stackoverflow.com/questions/25833613/safe-method-to-get-value-of-nested-dictionary
import functools 

def deep_get(dictionary, *keys):
    return functools.reduce(lambda d, key: d.get(key, None) if isinstance(d, dict) else None, keys, dictionary)
```

```python
data_file_subjectIDs = open('/Users/maria/Desktop/scopus/scopus_expanded_subjectIDs.csv', 'w', newline='')
csv_writer = csv.writer(data_file_subjectIDs)
 header = ["scopus_id", "subject_areas", "references"]
csv_writer.writerow(header)

count = 0

for scopus_id in scopus_ids_np_clean:
    query = "https://api.elsevier.com/content/abstract/scopus_id/"+str(scopus_id)+"?view=FULL"
    resp = requests.get(query,
                    headers={'Accept':'application/json',
                             'X-ELS-APIKey': MY_API_KEY})
    subject_areas = deep_get(resp.json(), "abstracts-retrieval-response", "subject-areas", "subject-area")
    references = deep_get(resp.json(), "abstracts-retrieval-response", "item", "bibrecord", "tail", "bibliography")
    csv_writer.writerow([scopus_id, str(subject_areas), str(references)])
    #if(count % 100 == 0):
    print(count)
    count = count+1
```
