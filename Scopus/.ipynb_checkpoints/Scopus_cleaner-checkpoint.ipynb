```python
from bs4 import BeautifulSoup
from langdetect import detect
import json
import os
```

```python
f = open('/Users/maria/Desktop/SCOPUS_api_search/scopus_1.json')
data = json.load(f)
```

```python
scopus_merged = list()
i=1
while(i<3290):
    filename='/Users/maria/Desktop/SCOPUS_api_search/scopus_'+str(i)+'.json'
    f = open(filename)
    data = json.load(f)
    scopus_merged = scopus_merged + data['search-results']['entry']
    i=i+1
```

```python
len(scopus_merged)
```

```python
scopus_no_dup_DOI = list()
bank = list()
for record in scopus_merged:
    if('prism:doi' in record.keys()):
        key = str(record['prism:doi'])
        if(key not in bank):
            bank.append(key)
            scopus_no_dup_DOI.append(record)
```

```python
len(scopus_no_dup_DOI)
```

```python
scopus_no_dup_DOI_abstract = list()
bank = list()
for record in scopus_no_dup_DOI:
    if('dc:description' in record.keys()):
        key = str(record['dc:description'])
        if(key not in bank):
            bank.append(key)
            scopus_no_dup_DOI_abstract.append(record)
```

```python
len(scopus_no_dup_DOI_abstract)
```

```python
scopus_no_dup_DOI_abstract_EN = list()
bank = list()
count = 0
for record in scopus_no_dup_DOI_abstract:
    desc = str((record['dc:description']))
    if(count % 100 == 0):
        print(count)
    count = count+1
    if(detect(desc)=='en'):
        scopus_no_dup_DOI_abstract_EN.append(record)
```

```python
scopus_no_dup_DOI_abstract_EN[1000]
```

```python
directory = "/Users/maria/Desktop/scopus"
os.chdir(directory)
i = 0
filename_stem = "scopus_abstract_"
for record in scopus_no_dup_DOI_abstract_EN:
    abstract = record['dc:description']
    filename = filename_stem+str(i)+'.txt'
    f = open(filename, "w")
    f.write(abstract)
    f.close()
    i = i+1 
```

```python
import csv
```

```python
len(scopus_no_dup_DOI_abstract_EN)
```

```python
data_file = open('/Users/maria/Desktop/scopus/scopus.csv', 'w', newline='')
csv_writer = csv.writer(data_file)
 
header = ["title", "abstract", "year", "citations-received", "decade", "journal", "scopus_id", "keywords"]
csv_writer.writerow(header)

count = 0

for data in scopus_no_dup_DOI_abstract_EN:
    if(int(str(data['prism:coverDate'])[0:4]) < 1970):
        decade = "pre-1970"
    elif(1970 <= int(str(data['prism:coverDate'])[0:4]) < 1980):
        decade = "1970-1979"
    elif(1980 <= int(str(data['prism:coverDate'])[0:4]) < 1990):
        decade = "1980-1989"
    elif(1990 <= int(str(data['prism:coverDate'])[0:4]) < 2000):
        decade = "1990-1999"
    elif(2000 <= int(str(data['prism:coverDate'])[0:4]) < 2010):
        decade = "2000-2009"
    elif(2010 <= int(str(data['prism:coverDate'])[0:4]) < 2020):
        decade = "2010-2019"
    elif(2020 <= int(str(data['prism:coverDate'])[0:4]) < 2030):
        decade = "2020-2029"
    else:
        decade = "other"
    csv_writer.writerow([data.get('dc:title'), data.get('dc:description'), str(data.get('prism:coverDate'))[0:4], data.get('citedby-count'), decade, data.get('prism:publicationName'), data.get('dc:identifier'), data.get('authkeywords')])
    if(count % 100 == 0):
        print(count)
    count = count+1
data_file.close()
```
